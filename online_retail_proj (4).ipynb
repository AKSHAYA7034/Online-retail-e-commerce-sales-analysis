# 2. **Data pre-processing**

## 2.2 Handling missing value

Several preprocessing techniques are used to guarantee the dataset is clean and trustworthy for analysis:

1. **Remove rows with missing CustomerID**  
   For customer-level analysis like RFM segmentation and behavioral insights, CustomerID is crucial. Rows without a CustomerID are eliminated since they cannot be connected to any customers.

2. **Replace missing values in the Description column**  
   Some goods don't have descriptions. These entries are filled with `"No Description"` to maintain consistency and avoid issues during grouping or text‑based analysis.

3. **Remove duplicate rows**  
   Sales totals, customer behavior metrics, and product performance analyses can all be distorted by duplicate records. To guarantee data accuracy, duplicate rows are found and eliminated after handling missing values.

These cleaning procedures are carried out by the code that follows.

```python
# Remove rows with missing Customer ID
df = df.dropna(subset=["Customer ID"])

# Fill missing descriptions
df["Description"] = df["Description"].fillna("No Description")
```

```python
#removing duplicates
df = df.drop_duplicates()
```

## 2.3 fixing incorrect data type

Correcting data types is an essential step in preparing the dataset for accurate analysis. Inaccurate aggregations, calculation mistakes, and problems with modeling or visualization might result from incorrectly formatted columns. In this step, the following actions are performed:

- **Convert date-related columns to datetime format**  
enables accurate time-based analysis of things like seasonality, monthly patterns, and customer behavior over time.

**Transform numeric columns that are stored as objects**  
  Some quantity or price fields may be incorrectly read as text due to formatting issues. Accurate computations and statistical analysis are made possible by converting them to numeric types.

- **Ensure categorical fields use appropriate data types**  
  Columns such as `Country`, `StockCode`, or `Description` may be better represented

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"], errors="coerce")
df["Customer ID"] = df["Customer ID"].astype("int64")
df["Quantity"] = pd.to_numeric(df["Quantity"], errors="coerce")
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")
```

## 2.4 Creating the TotalValue Column

**TotalValue** is a new functionality designed to analyze revenue at the transaction level.  
This column represents the monetary value of each line item in the dataset and is calculated by multiplying:


- **Quantity** – the number of units purchased  
The product's unit price, or **Price**  

Understanding sales performance, spotting high-value transactions, and assisting with additional studies like revenue trends, customer value segmentation, and product profitability all depend on this derived indicator.

The TotalValue for every row is calculated by the code below.

```python
df["TotalValue"] = df["Quantity"] * df["Price"]
```

## 2.5 Extracting Date and Time Components**

The `InvoiceDate` column is divided into distinct elements, such as **Year**, **Month**, **Day**, and **Hour**, to allow for thorough time-based analysis.  
This change facilitates a more thorough investigation of consumer behavior and sales trends, such as:


* Identifying **seasonal trends** and year‑over‑year changes
  
* Analysing **monthly sales performance** and peak periods  

* Understanding **daily purchasing behaviour**  

* Analyzing **hourly transaction patterns** to gain operational insights  

* Enabling time‑series visualisations and aggregations  

The dataset is made more adaptable for grouping, filtering, and producing time-based insights by removing certain elements.  
For every time component, the code below generates new columns.

```python
df["Year"] = df["InvoiceDate"].dt.year
df["Month"] = df["InvoiceDate"].dt.month
df["Day"] = df["InvoiceDate"].dt.day
df["Hour"] = df["InvoiceDate"].dt.hour
```

## 2.6 Monthly Sales Aggregation

To analyse revenue trends over time, the dataset is grouped by **Year** and **Month**, and the total sales value for each period is calculated. This stage aids in determining:

Overall sales trends throughout time  
* Seasonal or monthly fluctuations in revenue
  
* High‑performing and low‑performing months  
Long-term sales growth or fall  

We generate a structured time-series dataset that can be utilized for trend analysis, forecasting, and visualization by aggregating the `TotalValue` column on a monthly basis.  
The code below computes the total monthly revenue.

```python
monthly_sales = df.groupby(["Year", "Month"])["TotalValue"].sum().reset_index()
monthly_sales
```

## 2.7 Top 10 Best‑Selling Products

The dataset is grouped by **Description** and the total sales value for each product is computed in order to determine which products contribute the most to overall revenue. Summing the `TotalValue` column for every unique product allows us to identify:


* The items in the catalog with the highest earnings 

* Important sources of income for the company  

* Items that could need special attention when it comes to stock management  

Possibilities for focused advertising or marketing campaigns  

The top ten products are taken out of the results after the total revenue for each product has been calculated and sorted in descending order.  
The output below shows the ten most valuable products based on total sales.

```python
top_products = df.groupby("Description")["TotalValue"].sum().sort_values(ascending=False).head(10)
top_products
```
